```python
numeric_cols = ["Gls", "Ast", "PK", "Min"]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
print(df[["Gls", "Ast", "PK", "Min"]].dtypes)
```

```python
import numpy as np

np.random.seed(42)

# Rename goal column
df = df.rename(columns={"Gls ▼": "Gls"})

# Create realistic xG
df["xG"] = (
    df["Gls"] +
    np.random.normal(0, 2.5, len(df))
).clip(lower=0.1).round(2)

# Split players into groups
n = len(df)

overperformers = df.sample(frac=0.15, random_state=42).index
remaining = df.index.difference(overperformers)

underperformers = np.random.choice(
    remaining,
    size=int(0.15 * n),
    replace=False
)

average_players = df.index.difference(
    list(overperformers) + list(underperformers)
)

# Overperformers: goals > xG
df.loc[overperformers, "xG"] = (
    df.loc[overperformers, "Gls"] -
    np.random.uniform(1, 5, len(overperformers))
).clip(lower=0.1)

# Underperformers: goals < xG
df.loc[underperformers, "xG"] = (
    df.loc[underperformers, "Gls"] +
    np.random.uniform(1, 5, len(underperformers))
)

# Average finishers
df.loc[average_players, "xG"] = (
    df.loc[average_players, "Gls"] +
    np.random.uniform(-1, 1, len(average_players))
)

df["xG"] = df["xG"].round(2)

# xAG correlated with assists
df["xAG"] = (
    df["Ast"] +
    np.random.normal(0, 1.5, len(df))
).clip(lower=0).round(2)

# npxG+xAG
df["npxG+xAG"] = (
    (df["xG"] - df["PK"]) +
    df["xAG"]
).round(2)

# Progressive stats based on minutes played
minutes_factor = df["Min"] / df["Min"].max()

df["PrgC"] = (
    np.random.normal(40, 15, len(df)) *
    (0.5 + minutes_factor)
).clip(lower=5).astype(int)

df["PrgP"] = (
    np.random.normal(80, 25, len(df)) *
    (0.5 + minutes_factor)
).clip(lower=10).astype(int)

df["PrgR"] = (
    np.random.normal(60, 20, len(df)) *
    (0.5 + minutes_factor)
).clip(lower=5).astype(int)

# Round numeric columns
df["xAG"] = df["xAG"].round(2)

(df.head())
```

```python
# Make sure numeric columns are numeric
cols = ["90s", "Gls", "Ast", "G+A", "xG", "xAG", "npxG+xAG",
        "PrgC", "PrgP", "PrgR"]

for col in cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Avoid division by zero
df["90s"] = df["90s"].replace(0, np.nan)

# Create per-90 metrics
df["Gls_per90"] = (df["Gls"] / df["90s"]).round(2)
df["Ast_per90"] = (df["Ast"] / df["90s"]).round(2)
df["G+A_per90"] = (df["G+A"] / df["90s"]).round(2)

df["xG_per90"] = (df["xG"] / df["90s"]).round(2)
df["xAG_per90"] = (df["xAG"] / df["90s"]).round(2)
df["npxG+xAG_per90"] = (df["npxG+xAG"] / df["90s"]).round(2)

df["PrgC_per90"] = (df["PrgC"] / df["90s"]).round(2)
df["PrgP_per90"] = (df["PrgP"] / df["90s"]).round(2)
df["PrgR_per90"] = (df["PrgR"] / df["90s"]).round(2)
```

```python
df.head()
```

```python
from scipy.stats import percentileofscore

metrics = [
    "Gls_per90",
    "Ast_per90",
    "G+A_per90",
    "xG_per90",
    "xAG_per90",
    "npxG+xAG_per90",
    "PrgC_per90",
    "PrgP_per90",
    "PrgR_per90"
]

for metric in metrics:
    df[f"{metric}_pct"] = (
        df[metric]
        .rank(pct=True)
        .mul(100)
        .round(0)
        .astype(int)
    )
```

```python
from mplsoccer import Radar
```

```python
mbappe = df[df['Player'] == 'Kylian Mbappé']
mbappe.head()
```

```python
cols_to_plot = [
    "Gls_per90_pct",
    "Ast_per90_pct",
    "xG_per90_pct",
    "xAG_per90_pct",
    "npxG+xAG_per90_pct",
    "PrgC_per90_pct",
    "PrgP_per90_pct",
    "PrgR_per90_pct"
]
```

```python
radar = Radar(
              params= cols_to_plot,
              min_range = [0 for _ in cols_to_plot],
              max_range = [100 for _ in cols_to_plot]
              )
```

```python
fig,ax = radar.setup_axis()

rings_inner = radar.draw_circles(
    ax= ax,
    facecolor = 'white',
    edgecolor = 'blue',
    lw = 1, zorder = 1
)

radar_output = radar.draw_radar(
    ax = ax,
    values = list(mbappe[cols_to_plot].values.flatten()),
    kwargs_radar = {'facecolor' : '#77BA99'},
    kwargs_rings = {'facecolor' : '#EFF0D1'}
)

radar_labels = radar.draw_range_labels(
    ax = ax, fontsize = 15, fontproperties = 'monospace'
)

param_labels = radar.draw_param_labels(
    ax = ax, fontsize = 15, fontproperties = 'monospace'
)

ax.text(x = .5, y = 1.02, s = 'Kylian Mbappe', fontsize = 20, ha = 'center', va = 'center', transform = ax.transAxes)
```
