```python
import pandas as pd
import matplotlib.pyplot as plt
from highlight_text import fig_text

# Team colors
liv_col = "#C8102E"   # Liverpool Red
rm_col = "#00529F"    # Real Madrid Blue

fig, ax = plt.subplots(figsize=(12, 6))

for team in df['Squad'].unique():

    team_df = df[df['Squad'] == team].copy()

    # Add starting point
    team_df = pd.concat([
        pd.DataFrame({
            'Squad': [team],
            'Minute': [0],
            'xG': [0],
            'Outcome': [''],
            'cum_xG': [0],
            'half': [1]
        }),
        team_df
    ], ignore_index=True)

    # Add halftime point
    first_half_xg = team_df[team_df['half'] == 1]['cum_xG'].iloc[-1]

    team_df = pd.concat([
        team_df[team_df['half'] == 1],

        pd.DataFrame({
            'Squad': team,
            'Minute': 45,
            'xG': 0,
            'Outcome': '',
            'cum_xG': first_half_xg,
            'half': 2
        }, index=[0]),

        team_df[team_df['half'] == 2]
    ], ignore_index=True)

    color = liv_col if team == "Liverpool" else rm_col

    # Plot each half separately
    for half in team_df['half'].unique():

        half_df = team_df[team_df['half'] == half]

        ax.plot(
            half_df['Minute'],
            half_df['cum_xG'],
            drawstyle='steps-post',
            lw=3,
            color=color
        )

    # Goal markers
    goals = team_df[
        team_df['Outcome'].astype(str).str.contains(
            'goal',
            case=False,
            na=False
        )
    ]

    if len(goals) > 0:
        ax.scatter(
            goals['Minute'],
            goals['cum_xG'],
            s=180,
            zorder=5,
            color=color,
            edgecolor='black'
        )

    # Final xG label
    final_xg = team_df['cum_xG'].iloc[-1]

    ax.text(
        92,
        final_xg,
        f"{team} ({final_xg:.2f} xG)",
        color=color,
        fontsize=12,
        fontweight='bold',
        va='center'
    )

# Halftime line
ax.axvline(
    45,
    linestyle='--',
    color='gray',
    alpha=0.7
)

# Half labels
ax.text(
    22.5,
    -0.15,
    'First Half',
    ha='center',
    fontsize=11
)

ax.text(
    67.5,
    -0.15,
    'Second Half',
    ha='center',
    fontsize=11
)

# Grid
ax.grid(
    axis='y',
    linestyle='--',
    alpha=0.3
)

# Axis formatting
ax.set_xlim(0, 95)

ax.set_xticks([0, 15, 30, 45, 60, 75, 90])

ax.set_xlabel(
    "Match Minute",
    fontsize=12,
    fontweight='bold'
)

ax.set_ylabel(
    "Cumulative xG",
    fontsize=12,
    fontweight='bold'
)

# Remove clutter
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Title
fig_text(
    0.5,
    0.96,
    "<Real Madrid> 1-0 <Liverpool>\n2021/22 UEFA Champions League Final",
    ha="center",
    fontsize=18,
    fontweight="bold",
    highlight_textprops=[
        {"color": rm_col},
        {"color": liv_col}
    ]
)

# Subtitle
fig.text(
    0.5,
    0.90,
    "Cumulative Expected Goals (xG) Flow",
    ha="center",
    va='bottom',
    fontsize=11,
    color="gray"
)

plt.tight_layout(rect=[0, 0, 1, 0.88])

plt.show()
```
