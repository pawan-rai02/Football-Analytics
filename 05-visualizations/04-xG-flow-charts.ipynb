```python
df['cum_xG'] = df.groupby('Squad')['xG'].cumsum()
```

```python
df['Minute'] = pd.to_numeric(df['Minute'].astype(str).str.split('+').str[0])
df['half'] = df['Minute'].apply(lambda x: 1 if x <= 45 else 2)
```
