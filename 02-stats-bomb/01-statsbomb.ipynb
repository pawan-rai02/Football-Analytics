```python
all_matches = []

for match_id in sb.matches(competition_id = 43, season_id = 106).match_id:
    temp_match = sb.events(match_id = match_id)
    all_matches.append(temp_match)
```

```python
import pandas as pd
df = pd.concat(all_matches)
```
