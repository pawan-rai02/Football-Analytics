```python
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
df_raw = pd.read_csv('project2.csv')
```

```python
df = df_raw.copy(deep= True)
```

```python
df = df.dropna(subset= ['Score'])

df.shape
```

```python
df = df.drop(columns= ['Attendance', 'Time'])
```

```python
df.head()
```

```python
df[['home_goals', 'away_goals']] = df['Score'].str.split('–', expand= True).astype(int)
```

```python
df.head(
    
)
```

```python
def det_res(row):
    if row['home_goals'] > row['away_goals']:
        return 'Home win'
    
    elif row['home_goals'] < row['away_goals']:
        return 'Away win'

    else:
        return 'Draw'
```

```python
df['result'] = df.apply(det_res, axis= 1)
```

```python
df['result']
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Day'] = df['Date'].dt.day_name()
```

```python
df.head()
```

```python
df['season_start'] = df['Date'].apply(lambda x : x.year == 1 if x.month < 8 else x.year)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns= ['Day'])
```

```python
df.head()
```

```python
df.reset_index(drop = True, inplace= True)
```

```python
for x in df.Home.unique():
    
    temp_df = df[(df['Home'] == x) | (df['Away'] == x)]

    temp_df = temp_df.sort_values(['Date'])

    temp_df['goal_value_to_calc'] = temp_df.apply(lambda y : y['home_goals'] if y['Home'] == x else y['away_goals'], axis= 1)
    
    temp_df['rolling_avg_goals'] = temp_df['goal_value_to_calc'].rolling(window= 5, closed= 'left', min_periods= 1).mean()

    for index, row in temp_df.iterrows():
        if row['Home'] == x:
            df['home_rolling_avg_goals'] = pd.NA
            df['away_rolling_avg_goals'] = pd.NA

            for x in df.Home.unique():
                temp_df = df[(df['Home'] == x) | (df['Away'] == x)]
                temp_df = temp_df.sort_values(['Date'])

                temp_df['goal_value_to_calc'] = temp_df.apply(
                    lambda y: y['home_goals'] if y['Home'] == x else y['away_goals'],
                    axis=1
                )
                temp_df['rolling_avg_goals'] = temp_df['goal_value_to_calc'].rolling(
                    window=5, closed='left', min_periods=1
                ).mean()

                for index, row in temp_df.iterrows():
                    if row['Home'] == x:
                        df.at[index, 'home_rolling_avg_goals'] = row['rolling_avg_goals']
                    else:
                        df.at[index, 'away_rolling_avg_goals'] = row['rolling_avg_goals']

        else:
            df.at[index, 'away_rolling_avg_goals'] == row['rolling_avg_goals']
```

```python
for x in df.Home.unique():
    
    temp_df = df[(df['Home'] == x) | (df['Away'] == x)]

    temp_df = temp_df.sort_values(['Date'])

    temp_df['xG_value_to_calc'] = temp_df.apply(lambda y : y['xG'] if y['Home'] == x else y['xG.1'], axis= 1)
    
    temp_df['rolling_avg_xG'] = temp_df['xG_value_to_calc'].rolling(window= 5, closed= 'left', min_periods= 1).mean()

    for index, row in temp_df.iterrows():
        if row['Home'] == x:
            df['home_rolling_avg_xG'] = pd.NA
            df['away_rolling_avg_xG'] = pd.NA

            for x in df.Home.unique():
                temp_df = df[(df['Home'] == x) | (df['Away'] == x)]
                temp_df = temp_df.sort_values(['Date'])

                temp_df['xG_value_to_calc'] = temp_df.apply(
                    lambda y: y['xG'] if y['Home'] == x else y['xG.1'],
                    axis=1
                )
                temp_df['rolling_avg_xG'] = temp_df['xG_value_to_calc'].rolling(
                    window=5, closed='left', min_periods=1
                ).mean()

                for index, row in temp_df.iterrows():
                    if row['Home'] == x:
                        df.at[index, 'home_rolling_avg_xG'] = row['rolling_avg_xG']
                    else:
                        df.at[index, 'away_rolling_avg_xG'] = row['rolling_avg_xG']

        else:
            df.at[index, 'away_rolling_avg_xG'] == row['rolling_avg_xG']
```

```python
df = df.dropna(subset= ['home_rolling_avg_goals',
       'away_rolling_avg_goals', 'home_rolling_avg_xG', 'away_rolling_avg_xG'])
```
