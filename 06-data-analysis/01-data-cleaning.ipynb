```python
import pandas as pd

import numpy as np
```

```python
data = pd.DataFrame({
    'name' : ['Messi', 'Ronaldo', 'Neymar', 'Lingard', 'Mbappe', 'Messi'],
    'goals' : [45, 43, 32, 10.34, 39, 10],
    'assists' : [28, 14, 19, np.nan, 7, 20],
    'position' : ['Forward', 'Forward', 'Forward', 'Midfielder', 'FWD', 'FWD'],
    'birth-date' : ['1987-06-24 00:00:00', '1985-02-05', '1992-02-05', '1992-12-20', '1998-12-20', '1987-06-24 00:00:00']
})
```

```python
data
```

```python
data = pd.concat([data, data])
```

```python
data
```

```python
data = data.drop_duplicates()
```

```python
data
```

```python
data = data.drop_duplicates(subset= ['name'], keep = 'first')
data
```

```python
data['assists'] = data['assists'].fillna(0)
data
```

```python
data[['goals', 'assists']] = data[['goals', 'assists']].astype(int)
```

```python
data['position'] = data['position'].str.replace('FWD', 'Forward')
data
```

```python
data['birth-date']  = pd.to_datetime(data['birth-date'], format='mixed')
```
